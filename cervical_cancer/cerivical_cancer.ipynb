```python
df2=df[['Fold Change (linear) (patient vs. control)','ANOVA p-value (patient vs. control)','Gene Symbol']]
```

```python
df2['abs_log2']=np.log2(np.absolute(df['Fold Change (linear) (patient vs. control)']))
```

```python
df2['abs_log2'][df2['Fold Change (linear) (patient vs. control)']<0]=-(df2['abs_log2'][df2['Fold Change (linear) (patient vs. control)']<0])
```
