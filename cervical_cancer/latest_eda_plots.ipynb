```python
df.loc[(df['log2fold_change'] >= 1) & (df['ANOVA p-value (patient vs. control)'] < 0.05), 'color'] = "green"  # upregulated
df.loc[(df['log2fold_change'] <=- 1) & (df['ANOVA p-value (patient vs. control)'] < 0.05), 'color'] = "red"   # downregulated
df['color'].fillna('grey', inplace=True) # intermediate
```

```python
df = df.loc[(df['patient Bi-weight Avg Signal (log2)'] >= 10) & (df['control Bi-weight Avg Signal (log2)'] >= 10)]
```

```python
df['logpv']=-(np.log10(df['ANOVA p-value (patient vs. control)']))
```
